```python
import numpy as np
import pandas as pd
import re
```

```python
file_path = 'final_perfume_data.csv'
df = pd.read_csv(file_path,  encoding = 'unicode_escape')
```

```python
df.head(10)
```

```python

def normalize_text(text):
    text = re.sub(r'<.*?>', '', str(text))  
    text = text.encode("ascii", "ignore").decode() 
    text = re.sub(r'\s+', ' ', text) 
    return text.strip().lower()

df["Description"] = df["Description"].apply(normalize_text)
df["Notes"] = df["Notes"].apply(normalize_text)
df["final_description"] = (
    df["Description"].fillna('') + " " + df["Notes"].fillna('')
).str.strip()
df["final_description"] = df["final_description"].apply(normalize_text)
df = df.drop_duplicates(subset=["final_description"])
```

```python
df = df[df["final_description"].str.strip() != ""]
df = df.drop_duplicates(subset=["final_description"])
df = df[df["final_description"].str.len() > 20] 
```

```python
df.head(5)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df["final_description"])
```

```python
print(tfidf_matrix)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
print(cosine_sim)
```

```python
import numpy as np

np.fill_diagonal(cosine_sim, 0)

n = cosine_sim.shape[0]
max_score = 0
best_pair = (None, None)


for i in range(n):
    for j in range(i + 1, n):
            if cosine_sim[i, j] > max_score and cosine_sim[i, j] < 0.9 and df.iloc[i]["Brand"].lower() != df.iloc[j]["Brand"].lower():
                max_score = cosine_sim[i, j]
                best_pair = (i, j)

i, j = best_pair

print("Cele mai asemănătoare două parfumuri sunt:")
print("Parfum 1:", df.iloc[i]["Name"])
print("Parfum 2:", df.iloc[j]["Name"])
print("Scor de similaritate:", cosine_sim[i, j])


with open("README.txt", "w", encoding="utf-8") as f:
    f.write("Cele mai asemănătoare două produse din brand-uri diferite identificate:\n")
    f.write(f"Produs 1: {df.iloc[i]['Name']}\n")
    f.write(f"Descriere 1: {df.iloc[i]['Description']}\n")
    f.write(f"Notes 1: {df.iloc[i]['Notes']}\n")
    f.write(f"Produs 2: {df.iloc[j]['Name']}\n")
    f.write(f"Descriere 2: {df.iloc[j]['Description']}\n")
    f.write(f"Notes 2: {df.iloc[j]['Notes']}\n")
    f.write(f"Scor de similaritate: {cosine_sim[i, j]:.4f}\n")
```
